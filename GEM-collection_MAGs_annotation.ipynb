# Parsing the data from the terminal reductase annotation of the GEM metagenome data

Used Hmmer3 to annotate the GEM MAGs database for the following oxidoreductases using hmm models downloaded from InterPro. The GEM collection contains 52,515 metagenome-assembled genomes from over 10,000 metagenome and is publicly available.

- Cytochrome cbb3 oxidase (hmm accession PF00115)
- Cythocrome ubiquinol bd-I oxidase (PF01654)
- Particculate monoxigenase pMMO (PF02461)
- Nitrate reductase NarG (PF02665)
- Periplasmic nitrate reductase NapA (TIGR01706)
- Archaea ammonia monoxygenase AmoA (PF12942)
- Bacterial ammonia monoxygenase AmoA (NF041557)
- Dissimilatory sulfite reductase DsrA (TIGR02064)
- Sulfite reductase SoxC (TIGR04555)
- Thiosulfate reductase PshA (from a sulfur paper)

the data has been annotated using 

```hmmsearch --cpu 12 --tblout results_table_gene.txt MODEL.hmm.gz GEM_combined_mags.faa.gz > results_gene.log```

The steps to import and mathc the hmmer results to the mags metadata are:

- import the hmmer results in R using rhmmer
- export as plain csv
- remove the " characters using the sed command ```sed -i 's/"//g' input_file```
- run the pyhton script ```python3 match_files_hmmer.py gene_hmmer_result.csv mags_mapping_names.txt```
